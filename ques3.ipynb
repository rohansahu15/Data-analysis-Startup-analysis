```python
city = []
```

```python
for i in test3['CityLocation']:
    city.append(i)
```

```python
city
```

```python
np_city = np.array(city)
```

```python
np_city = np_city[np_city != '']
```

```python
for i in range(len(np_city)):
    if 'bangalore' in np_city[i]:
        np_city[i]='Bangalore'
    if np_city[i]=='Delhi':        
        np_city[i]='New Delhi'
```

```python
dic=dict()
for i in np_city:
    if i in dic.keys():
        dic[i]+=1
    else:
        dic[i]=1
```

```python
print(dic)
```

```python
for i in range(len(np_city)) :
        np_city[i]=np_city[i].split('/')[0].strip()
```

```python
dic=dict()
for i in np_city:
    if i in dic.keys():
        dic[i]+=1
    else:
        dic[i]=1
print(dic)
```

```python
x = []
y = []
```

```python
for i in dic:
    x.append(i)
    y.append(dic[i])
print(x,y)
```

```python
np_x = np.array(x)
np_y = np.array(y)
```

```python
np_x=np_x[np.argsort(np_y)]
np_y=np.sort(np_y)
print(np_x,np_y)
```

```python
np_x=np_x[len(np_x)-1:len(np_x)-1-10:-1]
np_y=np_y[len(np_y)-1:len(np_y)-1-10:-1]
```

```python
print(np_x,np_y)
```

```python
plt.pie(np_y,labels = np_x,autopct="%.2f%%",radius = 2)
```

```python
li = list(zip(np_x,np_y))
fundings = pd.DataFrame(li, columns = ["City", "No. Of Startups"])
fundings
```
